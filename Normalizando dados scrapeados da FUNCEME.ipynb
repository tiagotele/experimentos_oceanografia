```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Imputer

region = "Litoral De Fortaleza"
funceme_df = pd.read_csv("media_chuvas/funceme_media_macrorregiao.csv", index_col=0 ,parse_dates=['datahora'])
```

```python
funceme_df.loc['2012-08-01 12:00:00']
```

```python
media_observado_a_substituir = funceme_df.loc['2012-08-01 12:00:00']['Observado(mm)'].mean()
media_observado_a_substituir
```

```python
media_desvio_a_substituir = funceme_df.loc['2012-08-01 12:00:00']['Desvio(%)'].mean()
media_desvio_a_substituir
```

```python
funceme_df['Observado(mm)'].fillna(media_observado_a_substituir, inplace=True)
funceme_df['Desvio(%)'].fillna(media_desvio_a_substituir, inplace=True)
```

```python
medias = []
observado = []
desvio = []

for indices_unicos in funceme_df.index.unique():
    print(indices_unicos)
    print("Média normal ", funceme_df.loc[indices_unicos]['Normal(mm)'].mean())
    medias.append(funceme_df.loc[indices_unicos]['Normal(mm)'].mean())
    print("Observado ", funceme_df.loc[indices_unicos]['Observado(mm)'].mean())
    observado.append(funceme_df.loc[indices_unicos]['Observado(mm)'].mean())
    print("Desvio ", funceme_df.loc[indices_unicos]['Desvio(%)'].mean())
    desvio.append(funceme_df.loc[indices_unicos]['Desvio(%)'].mean())
```

```python
funceme_media_estadual_df = pd.DataFrame(index=funceme_df.index.unique().tolist())

```

```python

funceme_media_estadual_df = pd.concat(
    [funceme_media_estadual_df, 
        pd.DataFrame(data=medias, index=funceme_media_estadual_df.index, columns= ['Normal(mm)']),
        pd.DataFrame(data=observado, index=funceme_media_estadual_df.index, columns= ['Observado(mm)']),
        pd.DataFrame(data=desvio, index=funceme_media_estadual_df.index, columns= ['Desvio(mm)'])],
    axis=1, join_axes=[funceme_media_estadual_df.index])

```
